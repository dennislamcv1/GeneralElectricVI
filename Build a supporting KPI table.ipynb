Use a code recipe or visual recipe to INNER JOIN* the manufacturing tables to create a table that has the KPIs (Key Performance Indicators) of each part and the engine serial number (or ESN) they associate with.

```python
import numpy as np
import pandas as pd


#sets the default autosave frequency in seconds
%autosave 60 

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns',None)
#pd.set_option('display.max_rows',None)
pd.set_option('display.width', 1000)
np.random.seed(0)
np.set_printoptions(suppress=True)
```

```python
df = pd.read_csv("avenginedata.csv")
```

```python
df
```

```python
df1 = df.sample(frac=0.5,random_state=0)
```

```python
df1
```

### Step 5 | Build a supporting KPI table.

```python
manu = pd.read_csv("av_manufacturing_supply_chain_psql.csv")
```

```python
manu
```

```python
manu1 = manu.sample(frac=0.05, random_state=0)
```

```python
manu1
```

```python
bom = pd.read_csv("av_bom_manufacturing_psql.csv")
```

```python
bom
```

```python
bom1 = bom.sample(frac=0.1,random_state=0)
```

```python
kpitable = pd.merge(left=manu1, right=bom1, how='inner',on='pn')
```

```python
kpitable
```

```python
kpitable.duplicated().sum()
```

```python
kpitable1 = kpitable.sample(frac=0.05, random_state=0)
```

```python
kpitable1
```

### Step 6 | Join the new KPI table to the consolidated flights table

```python
step6 = pd.merge(left=df1, right=kpitable1, how='inner',on='esn')
```

```python
step6
```

### Step 7 | Join final table with table from step 5 to get remaining useful life (RUL*) for each engine

```python
rul = pd.read_csv("av_esn_rul_psql.csv")
```

```python
rul
```

```python
step7 = pd.merge(left=df1, right=rul, how='inner',on='esn')
```

```python
step7
```

```python
step7.to_csv("final.csv",index=False)
```
